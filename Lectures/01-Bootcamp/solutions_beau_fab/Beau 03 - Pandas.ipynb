```python
#====== 1 =======
print("Answer Q1: ")
data = pd.read_csv("../data/appl_1980_2014.csv") 
data.head()
```

```python
#====== 2 =======
print("Answer Q2: ")
data['Date'] = pd.to_datetime(data['Date'])
data.head()
```

```python
#====== 3 =======
print("Answer Q3: ")
morethan_1987 =  data['Date'] > '1986'
data = data[morethan_1987]
data.head()
```

```python
#====== 4 =======
print("Answer Q4: ")
data = data.set_index('Date')
data.sort_index(ascending=True, inplace = True)
data.head()
```

```python
#====== 5 =======
print("Answer Q5: ")
data_bymonth = data.resample('BM').mean()
data_bymonth.head()
```

```python
#====== 6 =======
print("Answer Q6: ")
index = data_bymonth.index
number_of_rows = len(index)
print(number_of_rows)
```

```python
#====== 7 =======
print("Answer Q7: ")
import matplotlib.pyplot as plt
data_high = data['High']
data_high.reset_index(drop=True, inplace=True)
data_high_shifted =data_high.shift(365)
diff = data_high - data_high_shifted
diff.plot(style=["r--"], label='diff')
plt.legend()
```

```python
#====== 8 =======
print("Answer Q8: ")
data_close = data['Close']
apple_close = data_close.reset_index(drop=True)
rolling = apple_close.rolling(100, center = True).mean()
apple_close.plot()
rolling.plot()
```

```python
#====== 9 =======
print("Answer Q9: ")
from pandas_datareader import data
microsoft = data.DataReader('MSFT', start='1987', end='2014', data_source='yahoo')
microsoft.head()
```

```python
microsoft_close = microsoft['Close']
microsoft_close = microsoft_close['2000':]

data_close = data_close['2000':]

microsoft_close.reset_index(drop=True, inplace=True)
data_close.reset_index(drop=True, inplace=True)
```

```python
ROI_microsoft = microsoft_close / microsoft_close[0]
ROI_apple = data_close/ data_close[0] 
```

```python
ROI_microsoft.plot(label="Microsoft")
ROI_apple.plot(label="Apple")
plt.legend()
```
